```python
import pandas as pd
import numpy as np
```

```python
dt=pd.read_csv("iris.csv")
dt
```

```python
sl=dt.iloc[:,0]
sw=dt.iloc[:,1]
pl=dt.iloc[:,2]
pw=dt.iloc[:,3]
s=dt.iloc[:,4]
```

```python
sl1=sl.tolist()
sw1=sw.tolist()
pl1=pl.tolist()
pw1=pw.tolist()
s1=s.tolist()
```

```python
from sklearn.model_selection import train_test_split
sltn,sltt,swtn,swtt,pltn,pltt,pwtn,pwtt,stn,stt=train_test_split(sl1,sw1,pl1,pw1,s1,train_size=0.8,random_state=101)
```

```python
c=0;
print("type found", "actual type",sep="\t")
for i in range(len(sltt)):
    m1=10000000;
    m2=10000001;
    m3=10000002;
    m4=10000003;
    s1,s2,s3,s4=" " ," ", " "," " 
    for j in range(len(sltn)):
        t1=(((sltn[j]-sltt[i])**2)+((pltn[j]-pltt[i])**2)+((swtn[j]-swtt[i])**2)+((pwtn[j]-pwtt[i])**2))
        if(t1<m4):
            if(t1<m3):
                m4=m3
                s4=s3
                if(t1<m2):
                    m3=m2
                    s3=s2
                    if(t1<m1):
                        m2=m1
                        s2=s1
                        m1=t1
                        s1=stn[j]
                    else:
                        m2=t1
                        s2=stn[j]
                else:
                    m3=t1
                    s3=stn[j]
            else:
                m4=t1
                s4=stn[j]
    
    c1,c2,c3=0,0,0
    if s1=="setosa":
        c1=c1+1;
    if s2=="setosa":
        c1=c1+1;
    if s3=="setosa":
        c1=c1+1;
    if s4=="setosa":
        c1=c1+1;
    
    if s4=="virginica":
        c2=c2+1;
    if s3=="virginica":
        c2=c2+1;
    if s2=="virginica":
        c2=c2+1;
    if s1=="virginica":
        c2=c2+1;
    
    if s4=="versicolor":
        c3=c3+1;
    if s3=="versicolor":
        c3=c3+1;
    if s2=="versicolor":
        c3=c3+1;
    if s1=="versicolor":
        c3=c3+1;
    if(c3>c2):
        if(c3>c1):
            st="versicolor"
        else:
            st="setosa"
    else:
        if(c2>c1):
            st="virginica"
        else:
            st="setosa"
    if st==stt[i]:
        c=c+1;
    if(st=="setosa"):
        print(st,stt[i],sep="\t\t")
    else:
        print(st,stt[i],sep="\t")

c=c*100/len(sltt)
print("Percentage accuracy ",c,"%")
```
