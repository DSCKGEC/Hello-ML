```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv("iris.csv")
#Importing csv as dataframe
data
```

```python
print(data)
```

```python
#Doing label encoding for species column
specie = data.pop('species')
print (specie)
for i in range(len(specie)):
    if specie[i]=='setosa':
        specie[i] = 0
    if specie[i]=='versicolor':
        specie[i] = 1
    if specie[i] == 'virginica':
        specie[i] = 2
print(specie)
data.insert(4,'species',specie)
```

```python
#Converting to numpy array
ar=data.to_numpy()
print(ar)
print(type(ar))
```

```python
#Dropping species column
newar=np.delete(ar, np.s_[4], axis=1)
print(newar)
```

```python
#Min max scaling
min_max_scaler = preprocessing.MinMaxScaler()
ar_minmax = min_max_scaler.fit_transform(newar)
ar_minmax
```
